# Region-Wise Attacks Distribution

Insights:
- **Most Attacked Region**: 
  - **Middle East & North Africa** reported the highest number of terrorist attacks.
- **Least Attacked Region**: 
  - **Australasia & Oceania** reported the fewest attacks.

**Implication**: This data can inform resource allocation for counterterrorism and humanitarian aid.


```python
region_wise_count = df.groupby('region.1')['region'].count().reset_index(name='count') 
plt.figure(figsize=(12,6))
ax = sns.barplot(x='region.1',y='count',data=region_wise_count,hue='region.1')
plt.title('Region Wise Attacks Distribution')
plt.xlabel('Region')
plt.ylabel('Number of Attacks')
for p in ax.containers:
    ax.bar_label(p, label_type='edge')
plt.xticks(rotation=90)
plt.show()
```

# Most Used Weapons in Attacks

Insights:
- **Most Used Attacks are Explosives**: 
- **Least Used Attacks are Radiological Weapons**: 

**Implication**: This data can inform resource allocation for counterterrorism and humanitarian aid.


```python
weapon_counts = df['weapon'].value_counts().reset_index()
plt.figure(figsize=(12,6))
ax = sns.barplot(
    weapon_counts,
    x='weapon',
    y='count',
    hue='weapon'
)
plt.title('Weapon Wise Attacks Distribution')
plt.xlabel('Weapon')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=75)
for p in ax.containers:
    ax.bar_label(p, label_type='edge')
plt.yscale('log')
plt.show()
```
